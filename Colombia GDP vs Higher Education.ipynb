## Colombia's GDP

```python
colombia_gdp = gdp_df.loc[gdp_df.Country_Name == 'Colombia']
colombia_gdp = colombia_gdp.drop(columns = ['Country_Name','Country_Code','Indicator_Name','mean'])
colombia_gdp
```

```python
colombia_years = []
for col in colombia_gdp.columns:
    colombia_years.append(col)
colombia_years
```

```python
colombia_percent = []

for index, row in colombia_gdp.iterrows():
    for i in range(0,len(colombia_gdp.columns)):
        colombia_percent.append(row[i])

colombia_percent
```

```python
colombia_gdp = pd.DataFrame({
    'Years': colombia_years,
    '% of GDP Growth': colombia_percent})
colombia_gdp
```

# Importing Stats about LATAM

```python
stats = pd.read_csv('LATAM_stats.csv', names=column_names)
stats
```

```python
stats = stats.loc[stats['Country_Name'].isin(latin_caribbean_countries)]
stats.reset_index(inplace=True)
stats.drop('index',axis = 1, inplace = True)
stats = stats[['Country_Name','Country_Code', 'Indicator_Name', 'Indicator_Code', '2011', '2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]
stats
```

### Colombia's School Enrollment

```python
colombia = stats.loc[stats.Country_Name == 'Colombia']
colombia
```

```python
colombia_enrollment = colombia[colombia['Indicator_Name'].str.match('^School enrollment.*')== True]
colombia_enrollment.reset_index(inplace=True)
colombia_enrollment
```

```python
colombia_enrollment_tertiary = colombia_enrollment.loc[2]
colombia_enrollment_tertiary
                                            
```

```python
colombia_enrollment_percent = []

for row in colombia_enrollment_tertiary:
    colombia_enrollment_percent.append(row)

colombia_enrollment_percent = colombia_enrollment_percent[5:]
colombia_enrollment_percent
```

```python
colombia_enrollment_tertiary = pd.DataFrame({
    'Years': colombia_years,
    '% of Enrollment': colombia_enrollment_percent})
colombia_enrollment_tertiary
```

# Colombia's GDP vs Enrollment

```python
sns.lineplot(data = colombia_gdp,
            x = "Years",
            y = '% of GDP Growth')
```

```python
sns.lineplot(data = colombia_enrollment_tertiary,
            x = 'Years',
            y = '% of Enrollment')
```

```python
# plot lines
plt.plot(colombia_gdp['Years'], colombia_gdp['% of GDP Growth'], label = "GDP")
plt.plot(colombia_enrollment_tertiary['Years'], colombia_enrollment_tertiary['% of Enrollment'], 
         label = "Higher Education Enrollment")

plt.legend()
plt.show()
```

Conclusion: There is surprisngly an inverse relationship between the % of GDP Growth in Colombia and % of individuals enrolled in higher education.
